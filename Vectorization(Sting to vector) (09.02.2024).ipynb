```python
from nltk.corpus import names
```

```python
def gender_features(name):
    return {"last letter":name[-1]}
```

```python
names.words()
```

```python
len(names.words())
```

```python
print(names.words('female.txt'))
```

```python
labeled_names=[(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')]
```

```python
labeled_names
```

```python
import random
random.shuffle(labeled_names)
```

```python
feature_sets=[(gender_features(n),gender) for (n,gender) in labeled_names]
len(feature_sets)
```

```python
train_set,test_set=feature_sets[5000:],feature_sets[:2944]
train_set
```

```python
import nltk
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features("Sathya"))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```

## Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer 
#from sklearn.feature_extraction.text import TfidVectorizer #this is good that count
```

```python
vect=CountVectorizer(binary=True)
corpus=["hello this is harini","Harini pursues computer science degree"]
vect.fit(corpus)
```

```python
vocab=vect.vocabulary_
print(vocab)

for i in sorted(vocab.keys):
    print()

```

```python
print(vect.fit_transform(corpus))
```

```python
print(vect.transform(corpus))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(vect.transform(["Corpus is insane amount of data","The degree is very important"]))
print(cs.tolist())
```
